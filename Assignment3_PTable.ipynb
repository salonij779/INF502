```python
def symbolInfo(element_symbol, lst):
    print("=======Information about " + element_symbol + " is as follows:=======")
    print("Atomic Number:" + str(lst[0]))
    print("Element Name:" + str(lst[1]))
    print("Symbol:" + str(lst[2]))
    print("Row:" + str(lst[3]))
    print("Column:" + str(lst[4]))
```

```python
elements = [[1, "Hydrogen", "H", 1, 1], [2, "Helium", "He", 1, 18], [3, "Lithium", "Li", 2, 1],
                [4, "Beryllium", "Be", 2, 2], [5, "Boron", "B", 2, 13], [6, "Carbon", "C", 2, 14]]

while True:
    print("\n========================= Start =========================\n")
    print("Enter your option from the following:")
    print("1.Information about any element, by entering that element's symbol.")
    print("2.Choose a property to see that property for all elements.")
    print("3.Enter a new element")
    print("4.Change the attributes of an element, by entering the element's symbol.")
    print("5.Exit the program")

    try:
        choice = int(input())
    except ValueError:
        choice = 0

    if choice == 0:
        print("\nKindly enter your choice.\n")
    elif choice == 1:
        count = 0
        symbol = str(input("Enter the symbol of the element:"))
        for row in elements:
            if symbol in row:
                print("========================= Choice 1 =========================")
                symbolInfo(symbol, row)
                break
            else:
                count += 1

        if count == len(elements):
                print("Data not found.")

    elif choice == 2:
        print("========================= Choice 2 =========================")
        print("Choose a property:")
        print("1.Atomic number")
        print("2.Element Name")
        print("3.Element Symbol")
        print("4.Row")
        print("5.Column")
        print("Default option 3.")

        try:
            choice2 = int(input())
        except ValueError:
            choice2 = 4

        if choice2 == 1:
            print("Listing atomic numbers of all elements in the list:")
            for row in elements:
                print(row[choice2-1])
        elif choice2 == 2:
            print("Listing all element names in the list:")
            for row in elements:
                print(row[choice2 - 1])
        elif choice2 == 3:
            print("Listing symbols of all elements in the list:")
            for row in elements:
                print(row[choice2 - 1])
        elif choice2 == 4:
            print("Listing row numbers of all elements in the list:")
            for row in elements:
                print(row[choice2 - 1])
        elif choice2 == 5:
            print("Listing column numbers of all elements in the list:")
            for row in elements:
                print(row[choice2 - 1])

    elif choice == 3:
        print("\n========================= Choice 3 =========================\n")
        print("Enter new element:")
        newlist = []
        atomicNumber = int(input("Enter atomic number (integer only):"))
        elementName = input("Element Name:")
        elementSymbol = input("Element Symbol:")
        rowNumber = int(input("Enter row number (integer only):"))
        columnNumber = int(input("Enter column number (integer only):"))
        newlist.append(atomicNumber)
        newlist.append(elementName)
        newlist.append(elementSymbol)
        newlist.append(rowNumber)
        newlist.append(columnNumber)

        elements.append(newlist)
        print("New data added.")

    elif choice == 4:
        print("\n========================= Choice 4 =========================\n")
        print("Choose a property to change:")
        print("1.Atomic number")
        print("2.Element Name")
        print("3.Row")
        print("4.Column")

        try:
            choice2 = int(input())
            sym = input("Enter the symbol:")
        except ValueError:
            choice2 = 0

        count = 0
        if sym is None:
            print("No data found.")
        elif choice2 == 1:
            UpdatedValue = input("Enter the updated atomic weight (integer only):")
            for row in elements:
                if sym == str(row[2]):
                    row[0] = int(UpdatedValue)
                    symbolInfo(sym, row)
                else:
                    count += 1
        elif choice2 == 2:
            UpdatedValue = input("Enter updated element name:")
            for row in elements:
                if sym == str(row[2]):
                    row[1] = UpdatedValue
                    symbolInfo(sym, row)
                else:
                    count += 1
        elif choice2 == 3:
            UpdatedValue = input("Enter updated row number (integer only):")
            for row in elements:
                if sym == str(row[2]):
                    row[3] = int(UpdatedValue)
                    symbolInfo(sym, row)
                else:
                    count += 1
        elif choice2 == 4:
            UpdatedValue = input("Enter updated column number (integer only):")
            for row in elements:
                if sym == str(row[2]):
                    row[4] = int(UpdatedValue)
                    symbolInfo(sym, row)
                else:
                    count += 1

        if count == len(elements):
            print("Data not found.")

    elif choice == 5:
        print("\n========================= Choice 5 =========================\n")
        print("========================= Exit =========================")
        exit(1)
    else:
        print("\nChoice entered is incorrect.\n")
```
